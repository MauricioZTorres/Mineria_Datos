## Data clustering
#### Create a model using k means and test it.

**Cambio de dataset pues los anteriores no eran muy buenos para clustering**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
mall = pd.read_csv('Mall_Customers.csv')
```

```python
mall.head()
```

## Iniciaremos con el dataset Mall_Customers

```python
#No es necesario dividir en un train/test set, pues k-means es un modelo no supervisado 
X1 = mall[['Annual Income (k$)','Spending Score (1-100)']] 
X1.head()
```

```python
#Para los cluster es recomendable standarizar los datos
scaler = StandardScaler()
X1_scaled = scaler.fit_transform(X1)
```

```python
#Importamos el modelo
kmeans = KMeans(n_clusters=5, random_state=32, n_init=10)
kmeans.fit(X1_scaled)

mall['Cluster'] = kmeans.labels_
```

```python
#Graficamos los resultados
plt.scatter(X1_scaled[:, 0], X1_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.title("K-means Clustering of Mall Customers")
plt.show()
```

```python
#La inercia es una medida relacionada a los clusters, y nos dice que tan juntos estan los datos de cada agrupacion
print("Inercia:", kmeans.inertia_)
```
